```python
import tabula
import pandas as pd
import numpy as np
```

```python
dataframe = tabula.read_pdf('C:/Users/Kunal/Desktop/Office/24 december pnb bank details/pnb.pdf',pages='all')
dataframe
```

```python
dataframe = dataframe.fillna(0)
dataframe.head()
```

```python
dataframe['Txn Date'] =pd.to_datetime(dataframe['Txn Date'])
dataframe.head()
```

```python
def converter(x):
    if type(x) == int:
        return float(x)
    else:
        if 'Dr.' in x:
            x = x.strip('Dr.')
        elif '\r1' in x:
            x = x.strip('\r1')
        elif ' Cr.' in x:
            x = x.strip(' Cr.')
        values = x.split(',')
        values = "".join(values)
        values = float(values)
        return values
```

```python
dataframe[['Dr Amount','Cr Amount','Balance']] = dataframe[['Dr Amount','Cr Amount','Balance']].applymap(converter)
```

```python
dataframe.index = dataframe['Txn Date']
dataframe=dataframe.drop(columns='Txn Date')
dataframe
```

```python
daily_debit_avg = dataframe.resample('D')['Dr Amount'].mean()
daily_debit_avg= pd.DataFrame({'Date':daily_debit_avg.index, 'Average':daily_debit_avg.values})
daily_debit_avg.to_csv('Daily_Debit_Average.csv')
daily_debit_avg
```

```python
daily_credit_avg = dataframe.resample('D')['Cr Amount'].mean()
daily_credit_avg= pd.DataFrame({'Date':daily_credit_avg.index, 'Average':daily_credit_avg.values})
daily_credit_avg.to_csv('Daily_Credit_Average.csv')
daily_credit_avg
```

```python
daily_balance_avg = dataframe.resample('D')['Balance'].mean()
daily_debit_avg= pd.DataFrame({'Date':daily_balance_avg.index, 'Average':daily_balance_avg.values})
daily_balance_avg.to_csv('Daily_Balance_Average.csv')
daily_balance_avg
```
