```python
import tabula
import pandas as pd
import numpy as np
```

```python
dataframe = tabula.read_pdf('C:/Users/Kunal/Desktop/Office/24 december pnb bank details/pnb.pdf',pages='all')
dataframe
```

```python
dataframe = dataframe.fillna(0)
dataframe.head()
```

```python
dataframe['Txn Date'] =pd.to_datetime(dataframe['Txn Date'])
dataframe.head()
```

```python
def converter(x):
    if type(x) == int:
        return float(x)
    else:
        if 'Dr.' in x:
            x = x.strip('Dr.')
        elif '\r1' in x:
            x = x.strip('\r1')
        elif ' Cr.' in x:
            x = x.strip(' Cr.')
        values = x.split(',')
        values = "".join(values)
        values = float(values)
        return values
```

```python
dataframe[['Dr Amount','Cr Amount','Balance']] = dataframe[['Dr Amount','Cr Amount','Balance']].applymap(converter)
```

```python
dataframe.index = dataframe['Txn Date']
dataframe=dataframe.drop(columns='Txn Date')
dataframe
```

```python
weekly_credit_avg = dataframe.resample('W')['Cr Amount'].mean()
weekly_credit_avg = pd.DataFrame({'Date':weekly_credit_avg.index, 'Average':weekly_credit_avg.values})
weekly_credit_avg.to_csv('weekly_credit_avg.csv')
weekly_credit_avg
```

```python
weekly_debit_avg = dataframe.resample('W')['Dr Amount'].mean()
weekly_debit_avg= pd.DataFrame({'Date':weekly_debit_avg.index, 'Average':weekly_debit_avg.values})
weekly_debit_avg.to_csv('Weekly_Debit_Average.csv')
weekly_debit_avg
```

```python
weekly_balance_avg = dataframe.resample('W')['Balance'].mean()
weekly_balance_avg= pd.DataFrame({'Date':weekly_balance_avg.index, 'Average':weekly_balance_avg.values})
weekly_balance_avg.to_csv('Weekly_Balance_Average.csv')
weekly_balance_avg
```
